### EmoLex 감성분석 
- EmoLex 감성 사전 :
    단어들을 긍정, 부정뿐만 아니라 공포, 기대, 신뢰, 놀람, 슬픔, 기쁨, 역겨움과 같은 8가지 감정으로 분류함. 
    인간의 정서 정보를 더욱 풍부하게 반영한다. 14,182개의 감성어를 보유하고 있다.
- EmoLex 감성 사전은 다음 링크에서 다운받을 수 있다. 
    논문, 다국어 감성사전 등을 포함한 자료라 파일은 100mb이나 영어 분석에 필요한 감성 사전은 2mb정도이다. 
    해당 사전의 파일 이름은 NRC-Emotion-Lexicon-Wordlevel-v0.92 이다. 버전에 따라 버전명은 달라질 수 있다.

- https://saifmohammad.com/WebPages/NRC-Emotion-Lexicon.htm 
- NRC 감성사전은 등록된 14,182개의 각 단어들에 대해 긍정, 부정 + 8가지의 감정 총 10가지 척도에 대해 0 또는 1로 라벨링되어있다. 
    따라서 해당 사전은 약 14만 1820개의 행으로 이루어져 있다. 하나의 단어마다 10개씩의 행으로 이루어진 데이터인 셈이다.

##### 'anger':분노 #'anticipation':기대 #'disgust':혐오 #'fear':두려움 #'joy':기쁨 #'sadness':슬픔 #'surprise':놀라움 #'trust':신뢰
#'negative': 14154, 
#'positive': 14154,

```python
NRC = pd.read_csv("data/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt",
                  engine="python", header=None, sep="\t")
NRC.head(20)
NRC.shape # (141540, 3)
NRC
```

```python
# 긍정, 부정 라벨은 삭제
mask = NRC[1].isin(['negative', 'positive'])
print(NRC[~mask].shape) # ~를 포함하게 되면 mask의 값을 제외
NRC = NRC[~mask] # (113232, 3)
```

```python
# 1로 라벨링 되어있는 라벨만 추출한다
NRC = NRC[(NRC != 0).all(1)]
# 인덱스 번호 리셋
NRC = NRC.reset_index(drop=True)
NRC.head(10)
NRC.tail(10) # 13871
```

```python
# comment 0번 행 텍스트 번역
comment = get_translate(apply_regular)
```

```python
# 정규표현식 문자분해
tokenizer = RegexpTokenizer('[\w]+')
# 불용어제거
stop_words = stopwords.words('english')
# 어간추출
p_stemmer = PorterStemmer()

# emolex분석할 텍스트 토큰화
raw = comment.lower()
tokens = tokenizer.tokenize(raw)
tokens
```

```python
stopped_tokens = [i for i in tokens if not i in stop_words] # 불용어 제거
match_words = [x for x in stopped_tokens if x in list(NRC[0])] # 사전과 매칭
match_words
```

```python
emotion=[]
for i in match_words: # 사전과 매칭된 단어 하나씩 불러오기
    temp = list(NRC.iloc[np.where(NRC[0] == i)[0],1]) #nrc 0열에서 i와 같은 단어 인덱스를 찾고 1열만 list로 가져온다
    # ex. star -> ['anticipation', 'joy', 'trust']
    for j in temp: # temp에 포함된 감정을 하나씩 추출하여 빈리스트에 추가
        emotion.append(j)

# 카운트 
sentiment_result1 = pd.Series(emotion).value_counts() #
print(sentiment_result1) # 시리즈
display(pd.DataFrame(sentiment_result1, columns=['count'])) # 데이터프레임
print(Counter(emotion)) # count 딕셔너리
```

```python
# 함수 적용하기위해 comment컬럼을 str로 변경
def change_to_str(x):
    return str(x)
# change_to_str(data1['comment'].iloc[0])

# apply 함수로 각 행마다 str로 변경하는 함수 적용
data1['comment'] = data1['comment'].apply(change_to_str)
data1['comment']
```

```python
# 정규표현식 함수 
def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')  # 한글 추출 규칙: 띄어 쓰기(1 개)를 포함한 한글
    result = hangul.sub('', text)  # 위에 설정한 "hangul"규칙을 "text"에 적용(.sub)시킴
#     result = result.replace(' ','')
    return result

# 각 comment마다 정규표현식만 추출
data1['comment'] = data1['comment'].apply(apply_regular_expression)
data1['comment']
```

```python
# 정규표현식으로만 추출된 comment에서 명사만 추출

okt = Okt()  # 명사 형태소 추출 함수
data1['comment_nouns'] = data1['comment'].apply(okt.nouns)

# 한국어 불용어 제거 : https://www.ranks.nl/stopwords/korean
stop_words = pd.read_csv("data/한국어불용어.txt", engine="python", header=None)
stop_words = stop_words[0].to_list() # 리스트로 변환

for n in range(0,len(data1['comment_nouns'])):
    # 명사추출된 단어에서 두글자 이상만 추출
    data1['comment_nouns'].iloc[n] = [x for x in data1['comment_nouns'].iloc[n] if len(x) > 1]
    # 한국어 불용어 제거
    data1['comment_nouns'].iloc[n] = [word for word in data1['comment_nouns'].iloc[n] if not word in stop_words]
    
data1['comment_nouns']
```
