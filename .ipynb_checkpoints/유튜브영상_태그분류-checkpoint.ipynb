```python
NRC = pd.read_csv("data/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt",
                  engine="python", header=None, sep="\t")
NRC.head(20)
NRC.shape # (141540, 3)
NRC
```

```python
# ê¸ì •, ë¶€ì • ë¼ë²¨ì€ ì‚­ì œ
mask = NRC[1].isin(['negative', 'positive'])
print(NRC[~mask].shape) # ~ë¥¼ í¬í•¨í•˜ê²Œ ë˜ë©´ maskì˜ ê°’ì„ ì œì™¸
NRC = NRC[~mask] # (113232, 3)
```

```python
# 1ë¡œ ë¼ë²¨ë§ ë˜ì–´ìˆëŠ” ë¼ë²¨ë§Œ ì¶”ì¶œí•œë‹¤
NRC = NRC[(NRC != 0).all(1)]
# ì¸ë±ìŠ¤ ë²ˆí˜¸ ë¦¬ì…‹
NRC = NRC.reset_index(drop=True)
NRC.head(10)
NRC.tail(10) # 13871
```

```python
# comment 0ë²ˆ í–‰ í…ìŠ¤íŠ¸ ë²ˆì—­
comment = get_translate(apply_regular)
```

```python
# ì •ê·œí‘œí˜„ì‹ ë¬¸ìë¶„í•´
tokenizer = RegexpTokenizer('[\w]+')
# ë¶ˆìš©ì–´ì œê±°
stop_words = stopwords.words('english')
# ì–´ê°„ì¶”ì¶œ
p_stemmer = PorterStemmer()

# emolexë¶„ì„í•  í…ìŠ¤íŠ¸ í† í°í™”
raw = comment.lower()
tokens = tokenizer.tokenize(raw)
tokens
```

```python
stopped_tokens = [i for i in tokens if not i in stop_words] # ë¶ˆìš©ì–´ ì œê±°
match_words = [x for x in stopped_tokens if x in list(NRC[0])] # ì‚¬ì „ê³¼ ë§¤ì¹­
match_words
```

```python
emotion=[]
for i in match_words: # ì‚¬ì „ê³¼ ë§¤ì¹­ëœ ë‹¨ì–´ í•˜ë‚˜ì”© ë¶ˆëŸ¬ì˜¤ê¸°
    temp = list(NRC.iloc[np.where(NRC[0] == i)[0],1]) #nrc 0ì—´ì—ì„œ iì™€ ê°™ì€ ë‹¨ì–´ ì¸ë±ìŠ¤ë¥¼ ì°¾ê³  1ì—´ë§Œ listë¡œ ê°€ì ¸ì˜¨ë‹¤
    # ex. star -> ['anticipation', 'joy', 'trust']
    for j in temp: # tempì— í¬í•¨ëœ ê°ì •ì„ í•˜ë‚˜ì”© ì¶”ì¶œí•˜ì—¬ ë¹ˆë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
        emotion.append(j)

# ì¹´ìš´íŠ¸ 
sentiment_result1 = pd.Series(emotion).value_counts() #
print(sentiment_result1) # ì‹œë¦¬ì¦ˆ
display(pd.DataFrame(sentiment_result1, columns=['count'])) # ë°ì´í„°í”„ë ˆì„
print(Counter(emotion)) # count ë”•ì…”ë„ˆë¦¬
```
